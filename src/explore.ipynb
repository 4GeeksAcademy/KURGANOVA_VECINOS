```python
import sqlite3

# Establecer la conexión a la base de datos
conn = sqlite3.connect("../data/movies_database.db")

# Crear una consulta SQL para unir las tablas
query = """
    SELECT *
    FROM movies_table
    INNER JOIN credits_table
    ON movies_table.title = credits_table.title;
"""

# Ejecutar la consulta SQL y obtener los resultados en un nuevo DataFrame
total_data = pd.read_sql_query(query, conn)

# Cerrar la conexión a la base de datos
conn.close()

# Eliminar columnas duplicadas, si las hay
total_data = total_data.loc[:, ~total_data.columns.duplicated()]

# Mostrar las primeras filas del DataFrame resultante
total_data.head()
```

Paso 3: Transforma los datos
Como puedes ver, hay algunas columnas con formato JSON. Selecciona, de cada uno de los JSONs, selecciona el atributo name y reemplaza las columnas genres y keywords. Para la columna cast, selecciona los tres primeros nombres.

```python
#read json and transform

def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value
```

```python
import pandas as pd
import json

# Suponiendo que las columnas JSON tienen nombres 'genres' y 'keywords'
# Función para extraer solo el atributo 'name' de cada objeto en el JSON
def extract_names_from_json(json_str):
    try:
        data = json.loads(json_str)
        names = [item['name'] for item in data]
        return names
    except:
        return None

# Aplicar la función a las columnas correspondientes
cleaned_data['genres'] = cleaned_data['genres'].apply(lambda x: extract_names_from_json(x) if pd.notnull(x) else None)
cleaned_data['keywords'] = cleaned_data['keywords'].apply(lambda x: extract_names_from_json(x) if pd.notnull(x) else None)

# Mostrar las primeras filas del DataFrame limpiado
cleaned_data.head()
```

```python
import pandas as pd
import json

# Suponiendo que la columna 'cast' contiene datos JSON
# Función para extraer los tres primeros nombres de la lista de diccionarios
def extract_first_three_names_from_json(json_str):
    try:
        data = json.loads(json_str)
        names = [item['name'] for item in data[:3]]
        return names
    except:
        return None

# Aplicar la función a la columna 'cast'
cleaned_data['cast'] = cleaned_data['cast'].apply(lambda x: extract_first_three_names_from_json(x) if pd.notnull(x) else None)

# Mostrar las primeras filas del DataFrame limpiado
cleaned_data.head()
```

Las únicas columnas que quedan por modificar son crew (equipo) y overview (resumen). Para la primera columna, transfórmala para que contenga el nombre del director. Para la segunda, conviértela en una lista.

```python
import pandas as pd
import json

# Suponiendo que la columna 'crew' contiene datos JSON y 'overview' es una cadena de texto
# Función para extraer el nombre del director de la columna 'crew'
def extract_director_from_crew(json_str):
    try:
        crew_data = json.loads(json_str)
        for member in crew_data:
            if member['job'] == 'Director':
                return member['name']
    except:
        return None

# Función para convertir la columna 'overview' en una lista
def convert_overview_to_list(overview_str):
    try:
        return [overview_str]
    except:
        return None

# Aplicar las funciones a las columnas correspondientes
cleaned_data['crew'] = cleaned_data['crew'].apply(lambda x: extract_director_from_crew(x) if pd.notnull(x) else None)
cleaned_data['overview'] = cleaned_data['overview'].apply(lambda x: convert_overview_to_list(x) if pd.notnull(x) else None)

# Mostrar las primeras filas del DataFrame limpiado
cleaned_data.head()
```

Una vez hayamos terminado de procesar las columnas y que el modelo de recomendación no se confunda, por ejemplo, entre Jennifer Aniston y Jennifer Conelly, quitaremos los espacios entre las palabras. Aplica esta función a las columnas genres, cast, crew y keywords.

```python
#transform to a one single column

cleaned_data["tags"] = cleaned_data["overview"] + cleaned_data["genres"] + cleaned_data["keywords"] + cleaned_data["cast"] + cleaned_data["crew"]
cleaned_data["tags"] = cleaned_data["tags"].apply(lambda x: ",".join(x).replace(",", " "))

cleaned_data.drop(columns = ["genres", "keywords", "cast", "crew", "overview"], inplace = True)

#cleaned_data.iloc[0].tags
```
