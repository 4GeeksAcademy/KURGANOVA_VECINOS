Paso 2: Creación de una base de datos
Crea una base de datos para almacenar los dos DataFrames en tablas distintas. A continuación, une las dos tablas con SQL (e intégralo con Python) para generar una tercera tabla que contenga información de ambas unificada. La clave a través de la cual se puede hacer la unión es el título de la película (titulo).

```python
import sqlite3

conn = sqlite3.connect("../data/movies_database.db")

# Guardar los DataFrames en tablas SQLite
movies.to_sql("movies_table", conn, if_exists = "replace", index = False)
credits.to_sql("credits_table", conn, if_exists = "replace", index = False)
```

```python
import sqlite3

# Establecer la conexión a la base de datos
conn = sqlite3.connect("../data/movies_database.db")

# Crear una consulta SQL para unir las tablas
query = """
    SELECT *
    FROM movies_table
    INNER JOIN credits_table
    ON movies_table.title = credits_table.title;
"""

# Ejecutar la consulta SQL y obtener los resultados en un nuevo DataFrame
total_data = pd.read_sql_query(query, conn)

# Cerrar la conexión a la base de datos
conn.close()

# Eliminar columnas duplicadas, si las hay
total_data = total_data.loc[:, ~total_data.columns.duplicated()]

# Mostrar las primeras filas del DataFrame resultante
print(total_data.head())
```
